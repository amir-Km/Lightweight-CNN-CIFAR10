```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, callbacks, backend
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# ---
# 1. Data Preparation
# ---

# Load CIFAR-10 dataset
(x_train_full, y_train_full), (x_test, y_test) = cifar10.load_data()

# Normalize pixel values to be between 0 and 1
x_train_full = x_train_full.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Convert labels to one-hot encoding
y_train_full = to_categorical(y_train_full, 10)
y_test = to_categorical(y_test, 10)

# Extract the last 5000 samples for validation (as requested)
val_count = 5000
x_train = x_train_full[:-val_count]
y_train = y_train_full[:-val_count]
x_val = x_train_full[-val_count:]
y_val = y_train_full[-val_count:]

# ---
# 2. Advanced Model Architecture (< 400k parameters)
# ---

# Squeeze-and-Excitation (SE) Block for attention mechanism
def se_block(input_tensor, ratio=16):
    filters = input_tensor.shape[-1]
    se = layers.GlobalAveragePooling2D()(input_tensor)
    se = layers.Reshape((1, 1, filters))(se)
    se = layers.Dense(filters // ratio, activation='relu', use_bias=False)(se)
    se = layers.Dense(filters, activation='sigmoid', use_bias=False)(se)
    return layers.Multiply()([input_tensor, se])

# Modern Swish activation function
def swish_activation(x):
    return x * backend.sigmoid(x)

# Custom Residual Block with Separable Convolutions (to reduce parameters)
def residual_block(x, filters, stride=1):
    shortcut = x

    x = layers.SeparableConv2D(filters, (3, 3), strides=stride, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation(swish_activation)(x)

    x = layers.SeparableConv2D(filters, (3, 3), padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)

    # Inject Attention (SE Block)
    x = se_block(x)

    # Identity mapping (Skip connection)
    if stride != 1 or shortcut.shape[-1] != filters:
        shortcut = layers.Conv2D(filters, (1, 1), strides=stride, padding='same', use_bias=False)(shortcut)
        shortcut = layers.BatchNormalization()(shortcut)

    x = layers.Add()([x, shortcut])
    x = layers.Activation(swish_activation)(x)
    return x

# Build the final SE-ResNet model
def create_se_resnet():
    inputs = layers.Input(shape=(32, 32, 3))

    x = layers.Conv2D(32, (3, 3), padding='same', use_bias=False)(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation(swish_activation)(x)

    x = residual_block(x, 64)
    x = residual_block(x, 64)
    x = layers.Dropout(0.2)(x)

    x = residual_block(x, 128, stride=2)
    x = residual_block(x, 128)
    x = layers.Dropout(0.3)(x)

    x = residual_block(x, 256, stride=2)
    x = residual_block(x, 256)
    x = layers.Dropout(0.4)(x)

    # Global pooling to eliminate dense layer parameter explosion
    x = layers.GlobalAveragePooling2D()(x)
    outputs = layers.Dense(10, activation='softmax')(x)

    return models.Model(inputs=inputs, outputs=outputs)

model = create_se_resnet()
model.summary() # Verification: Should be ~392k parameters

# ---
# 3. Model Training
# ---

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Data Augmentation to prevent overfitting
datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
)
datagen.fit(x_train)

# Callbacks for dynamic learning rate and early stopping
checkpoint = callbacks.ModelCheckpoint("best_se_resnet.keras", monitor='val_accuracy', save_best_only=True, mode='max')
reduce_lr = callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, min_lr=1e-6)
early_stop = callbacks.EarlyStopping(monitor='val_accuracy', patience=15, restore_best_weights=True)

print("Starting training process...")
history = model.fit(datagen.flow(x_train, y_train, batch_size=64),
                    epochs=85,
                    validation_data=(x_val, y_val),
                    callbacks=[checkpoint, reduce_lr, early_stop])

# ---
# 4. Evaluation and Results
# ---

# Plot Training History
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.legend()
plt.show()

# Final Evaluation on unseen test set
loss, acc = model.evaluate(x_test, y_test)
print(f"\nğŸš€ Final Test Accuracy: {acc*100:.2f}%")
```
